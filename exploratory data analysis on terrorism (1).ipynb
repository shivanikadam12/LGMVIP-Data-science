```python
new_data = data[['eventid','iyear','imonth','country','country_txt','region','region_txt','provstate','city','crit1','crit2','success','suicide','attacktype1','attacktype1_txt','targtype1','targtype1_txt','natlty1','gname','guncertain1','claimed','weaptype1','nkill','nwound']]
new_data.head()
```

```python
new_data.isnull().sum()
```

# merging and comparing feature

```python
new_data['nkill'] = new_data['nkill'] .fillna(0) 
new_data['nwound'] = new_data['nwound'].fillna(0) 
new_data['casualities']=new_data['nkill']+ new_data ['nwound']
new_data.isnull().sum()
```

```python
new_data.describe()
```

```python
new_data.info()
```

```python
#analysis of visualization of data
```

```python
#year wise attack
```

```python
year=new_data['iyear'].unique()
year_order = new_data['iyear'].value_counts(dropna=False).sort_index()
plt.figure(figsize=(20,10))
sns.barplot(x=year,y=year_order,palette="Paired")
plt.xticks(rotation=90) 
plt.xlabel("Year")
plt.ylabel("No of Attacks")
plt.title("No of attacks year wise")
plt.show()
```

```python
#year wise casualiities
```

```python
cas_year = new_data[['iyear','casualities']].groupby('iyear').sum()
cas_year.plot(kind= 'bar', figsize=(20,10))
plt.title('Casualities year wise')
plt.xlabel("Years")
plt.ylabel("No of casualities")
plt.show()
```

```python
#types of attack
```

```python
new_data["attacktype1"].value_counts().plot(kind="bar",figsize=(20,10))
plt.xticks(rotation=0)
plt.xlabel("Attack Type")
plt.ylabel("No of attacks")
plt.title ("Types of Attacks")
plt.show()
```

```python
#top affected countries
```

```python

plt.subplots(figsize=(20,10))
sns.barplot(new_data['country_txt'].value_counts()[:10].index,new_data['country_txt'].value_counts()[:10].values)
plt.title('Top affected countires')
plt.xlabel('Countries')
plt.ylabel('Frequency-Count')
plt.xticks(rotation=50)
plt.show()
```

```python
#top affected cities
```

```python
new_data['city'].value_counts().to_frame().drop("Unknown").sort_values("city",axis=0,ascending=False).head(10).plot(kind='bar',figsize=(20,10))
plt.xticks(rotation = 50)
plt.xlabel("City")
plt.ylabel("No of attack") 
plt.title("Top 10 most effected city") 
plt.show()
```

```python
#top terrorist group
```

```python

new_data['gname'].value_counts().to_frame().drop("Unknown").head(10).plot(kind="bar",figsize=(20,10))
plt.xticks(rotation=70) 
plt.xlabel('Group Name')
plt.ylabel('No of attacks') 
plt.title('Top Terrorist Group')
plt.show()
```

```python
#type of killing
```

```python
plt.subplots(figsize=(20,10))
sns.countplot(new_data["targtype1_txt"], order=new_data['targtype1_txt'].value_counts().index);
plt.xticks(rotation=90) 
plt.xlabel("Target type")
plt.ylabel("count")
plt.title("Target Type")
plt.show()
```

```python
#terrorist attack region wise each year
```

```python
pd.crosstab(new_data.iyear, new_data.region_txt).plot(kind="area",stacked=False,figsize=(20,10))
plt.title("Terrorist Attacks Region wise in each year")
plt.ylabel("No of attacks") 
plt.xlabel("Year")
plt.show()
```
